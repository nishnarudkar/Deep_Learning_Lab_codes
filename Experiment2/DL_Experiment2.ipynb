```python
import numpy as np
from sklearn.linear_model import Perceptron
```

### AND Gate

```python
X_and = np.array([[0,0],[0,1],[1,0],[1,1]])
y_and = np.array([0,0,0,1])
```

```python
model_and = Perceptron(max_iter = 1000 , tol = 1e-3)
model_and.fit(X_and,y_and)
```

```python
print(f"AND Predictions {X_and}")
```

### OR GATE


```python
X_or = np.array([[0,0],[0,1],[1,0],[1,1]])
y_or = np.array([0,0,0,1])
```

```python
model_or = Perceptron(max_iter = 1000, tol = 1e-3)
model_or.fit(X_or,y_or)
```

```python
print(f"OR Predictions {X_or}")
```

### NOT GATE

```python
X_not = np.array([[0], [1]])
y_not = np.array([1, 0])

model_not = Perceptron(max_iter=1000, tol=1e-3)
model_not.fit(X_not, y_not)

print("NOT Predictions:", model_not.predict(X_not))
```

### Use Iris Dataset - Setosa vs Versicolor

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
iris = load_iris()
X = iris.data[:, :2]
y = iris.target
```

```python
X = X[y != 2]
y = y[y != 2]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

model = Perceptron(max_iter=1000, tol=1e-3)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```
